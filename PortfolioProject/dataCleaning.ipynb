```python
#Import packages
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#Import the dataset
home = pd.read_excel('homeLoanApproval.xlsx')

#Drop column unnecessary column 
home = home.drop(columns='Loan_ID')

#Number of instances and columns(614, 12)
home.shape

#Seperating features into categorical and numerical
cat_features = ['Gender', 'Married', 'Dependents', 'Education', 'Self_Employed', 'Property_Area','Credit_History','Loan_Amount_Term']

num_features =  ['ApplicantIncome', 'CoapplicantIncome', 'LoanAmount']

#Display graphs of categorialcal features with comparing approved and denied applicants
fig,axes = plt.subplots(4,2,figsize=(12,15))
for idx,cat_col in enumerate(cat_features):
    row,col = idx//2,idx%2
    sns.countplot(x=cat_col,data=home,hue='Loan_Status',ax=axes[row,col])

plt.subplots_adjust(hspace=1)

#Displays boxplot of numerical value displays outliers for each feature
fig,axes = plt.subplots(1,3,figsize=(17,5))
for idx,cat_col in enumerate(num_features):
    sns.boxplot(y=cat_col,data=home,x='Loan_Status',ax=axes[idx])

#Convert categorical features into indicator variables(0 or 1)
home_encoded = pd.get_dummies(home, drop_first=True)

#Drop all null values from data set
home_encoded = home_encoded.dropna()

#Save modified data set to new excel spread sheet
home_encoded.to_excel('Modified.xlsx')

#Display new dataset in terminal
home_encoded












 









```
