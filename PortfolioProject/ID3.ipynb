```python
#Import packages
import pandas as pd
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report

#Import modified library
hLoan = pd.read_excel('Modified.xlsx')

#Setting X to all descriptive features except Loan_Status_Y
X = hLoan.drop(columns= 'Loan_Status_Y')

#Setting y to Target Feature
y = hLoan['Loan_Status_Y']

#Splitting Dataset into Test 20% and Train 80% 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.2, random_state=1)

# Function to perform training and fit features
clf_en = DecisionTreeClassifier(criterion='entropy', max_depth=3, random_state=0)
clf_en.fit(X_train, y_train)

#Use model to make predictions on test data
y_pred = clf_en.predict(X_test)

#Print Classification Report for model 
print(classification_report(y_test, y_pred))

#Displays tree
plt.figure(figsize=(12,8))
tree.plot_tree(clf_en.fit(X_train, y_train)) 

```
